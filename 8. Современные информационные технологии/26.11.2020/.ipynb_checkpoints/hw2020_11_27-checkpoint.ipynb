# SSA. Гусеница

Импортируем библиотеки

```python
import datetime as dt
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.decomposition import PCA
```

## Работа с рядом Fort

Импортируем данные

```python
parser = lambda s: dt.datetime.strptime(s, '%b %Y')
```

```python
data = pd.read_csv('wine.csv', sep=';', index_col='date_', parse_dates=['date_'], date_parser=parser)
data = data.drop(columns=['year_', 'month_'])
data.head()
```

Посмотрим на картинки

```python
data.plot(figsize=(10, 6))
plt.show()
```

Стационарным ни один из временных рядов не назовёшь. Тем не менее, тренд отсутствует, ярко выражена годовая сезонность. 

```python
data['fort'].plot(figsize=(10, 6))
plt.title('Fort')
plt.show()
```

Посмотрим на временной ряд `Fort`. Длина временного ряда: 

```python
data['fort'].shape
```

Длину окна рекомендуется выбирать близкой к половине ряда. Половина ряда есть ничто иное, как

```python
174/2
```

Но мы возьмём $L=84$, т. к. 84 делится на 2, 3, 4, 6, 7, 12, 14, 21, 28, 42, а мы предполагаем, что у нас есть сезонность по меньшей мере в 12 и 6 месяцев (видно визуально). 

```python
N = 174
L = 84
K = N - L + 1
```

Составляем траекторную матрицу. Для этого нам понадобится вспомогательная функция `partition`. 

```python
def partition(lst, n, d):
    indexes_1 = np.arange(n)
    indexes_2 = np.arange(0, len(lst) - n + 1, d)
    return lst[indexes_1[np.newaxis] + indexes_2[:, np.newaxis]]
```

```python
ts = data['fort']
```

```python
trajectory_matrix = partition(ts.values, L, 1)
print(trajectory_matrix)
```

```python
plt.matshow(trajectory_matrix)
plt.show()
```

Проводим `PCA`

```python
pca = PCA().fit(trajectory_matrix)
```

```python
Y = pca.transform(trajectory_matrix)
```

Посмотрим на главные компоненты

```python
fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(16, 9))

for i in range(10):
    axes[i//2, i%2].plot(np.arange(91), Y[:, i], label=str(i))
    
plt.legend()
plt.show()
```

Скорее всего, третья главная компонента отвечает за тренд. Первая и вторая – за гармоники годовой сезонности, четвёртая и пятая – сезонность два месяца, и так далее. 

## Работа с другими исходными данными

```python
parser = lambda s: dt.datetime.strptime(s, '%Y-%m')
```

```python
data = pd.read_csv('monthly-us-female-1619-years-une.csv', index_col=0, parse_dates=[0], date_parser=parser)
data.head()
```

```python
data.plot(figsize=(10, 6))
plt.show()
```

В ряде явно присутствуют тренд и сезонность. Попробуем применить SSA к этому ряду. Длина ряда:

```python
ts = data['Monthly U.S. female (16-19 years) unemployment figures (thousands) 1948-1981']
ts.shape
```

Ширину окна возмьмём $L = 202$

```python
N = 408
L = 202
K = N - L + 1
```

```python
trajectory_matrix = partition(ts.values, L, 1)
print(trajectory_matrix)
```

```python
plt.matshow(trajectory_matrix)
plt.show()
```

Проводим `PCA`

```python
pca = PCA().fit(trajectory_matrix)
```

```python
Y = pca.transform(trajectory_matrix)
```

Видим, что вервая компонента отвечает на тренд, восьмая – за структурные изменения, между ними есть компоненты, которые отвечают за сезонность.
