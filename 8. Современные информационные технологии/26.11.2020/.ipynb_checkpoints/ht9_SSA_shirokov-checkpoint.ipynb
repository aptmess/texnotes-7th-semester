## 1. SSA Realisation

Пусть $N>2$. Рассмотрим вещественнозначный временной ряд $F = (f_0, \ldots, f_{N-1})$ длины N

### Шаг 1. Вложение

```python
F = wine.total.values
```

```python
N = F.shape[0]
N 
```

Пусть есть некоторое целое число `L` (длина окна): $1 < L < N$. 

Процедура вложения образует `K = N - L + 1` **векторов вложения**

$$\large X_i = (f_{i-1}, \ldots, f_{i+L-2})^T, \quad 1 \leqslant i \leqslant K$$

имеющих размерность $L$

```python
N / 2
```

```python
L = 84
```

```python
K = N - L + 1
K
```

$L$ - траекторная матрица (траекторная матрица ряда $F$):

$$\large X = [X_1: ... : X_k]$$

Другими словами, траекторая матрица - это матрица:

$$\large X \in \mathbb{R}^{L, k} = (x_{ij})_{i,j=1}^{L, K} = \begin{pmatrix} 
f_0 & f_1 & f_2 & \ldots & f_{K-1} \\ 
f_1 & f_2 & f_3 & \ldots & f_{K} \\
\vdots & \vdots & \vdots & \vdots & \vdots \\ 
f_{L-1} & f_{L} & f_{L+1} & \ldots & f_{N-1} \\
\end{pmatrix}$$

```python
X = scipy.linalg.hankel(c=F[0:L], 
                        r=F[L-1:])
#X_1 = np.column_stack([F[i:i+L] for i in range(0,K)]) или так
X.shape
```

```python
ax = plt.matshow(X)
plt.xlabel("$K$")
plt.ylabel("$L$")
plt.colorbar(ax.colorbar, fraction=0.025)
ax.colorbar.set_label("$F$")
plt.title("Траекторная матрица");
```

## Шаг 2. Сингулярное разложение

```python
U, Sigma, V = scipy.linalg.svd(X)
V = V.T
d = np.linalg.matrix_rank(X)
```

## Шаг 3. Групировка собственных троек

```python
X_elem = np.array([Sigma[i] * np.outer(U[:,i], V[:,i]) for i in range(0, d)])
```

```python
if not np.allclose(X, X_elem.sum(axis=0), atol=1e-10):
    print("Предуреждение: SVD проделано неверно..")
```

## Шаг 4. Диагональное усреднение

```python
def diag_mean(Y):
    L = min(Y.shape)
    K = max(Y.shape)
    N = L + K - 1
    if L < K:
        Y_ = Y
    elif L >= K:
        Y_ = Y.T
    g_1 = np.array([1 / (k + 1) * np.sum([Y_[m, k-m] for m in range(k+1)]) for k in range(0, L-1)])
    g_2 = np.array([1 / (L-1) * np.sum([Y_[m, k-m] for m in range(0, L-1)]) for k in range(L-1, K)])
    g_3 = np.array([1 / (N-k) * np.sum([Y_[m, k-m] for m in range(k-K+1, N-K+1)]) for k in range(K, N)])
    return np.r_[g_1,g_2,g_3]
```

## 2. `class SingularSpectrumAnalysis`

```python
import numpy as np
import scipy
class SingularSpectrumAnalysis():
    def __init__(self, F):
        '''
        F: np.ndarray
            input data
        '''
        self.F = F
        
    def step_1(self, L):
        '''
        L: int
            ширина окна
        '''
        X = scipy.linalg.hankel(c=self.F[:L], 
                                r=self.F[L-1:])
        return X
    
    def step_2(self, X):
        '''
        X: np.ndarray L x K
            траекторная матрица
        '''
        U, Sigma, V = scipy.linalg.svd(X)
        V = V.T
        d = np.linalg.matrix_rank(X)
        return U, Sigma, V, d
    
    def step_3(self, U, Sigma, V, d, X):
        '''
        группировка собственных троек
        '''
        X_elem = np.array([Sigma[i] * np.outer(U[:,i], V[:,i]) for i in range(0, d)])
        if not np.allclose(X, X_elem.sum(axis=0), atol=1e-10):
            print("Предуреждение: SVD проделано неверно..")
        return X_elem
    
    def step_4(self, Y):
        '''
        Y: np.ndarray
        
        Output:
            возвращает диагональное усреднение
        '''
        L = min(Y.shape)
        K = max(Y.shape)
        N = L + K - 1
        if L < K:
            Y_ = Y
        elif L >= K:
            Y_ = Y.T
        g_1 = np.array([1 / (k + 1) * np.sum([Y_[m, k-m] for m in range(k+1)]) for k in range(0, L-1)])
        g_2 = np.array([1 / (L-1) * np.sum([Y_[m, k-m] for m in range(0, L-1)]) for k in range(L-1, K)])
        g_3 = np.array([1 / (N-k) * np.sum([Y_[m, k-m] for m in range(k-K+1, N-K+1)]) for k in range(K, N)])
        return np.r_[g_1,g_2,g_3]
        
    def fit(self, n_components, L='default'):
        '''
        n_components:
            количество рядов
        '''
        N = self.F.shape[0]
        if L == 'default':
            L = np.int(N / 2)
        if n_components > L:
            raise ValueError(f'n_components = {n_components} should be less, than L = {L}')
        K = N - L + 1
        X = self.step_1(L)
        U, Sigma, V, d = self.step_2(X)
        X_elem = self.step_3(U, Sigma, V, d, X)
        self.g = np.array([self.step_4(x) for x in X_elem])
        return self.g[:n_components, :].sum(axis=0)
```

```python
SSA = SingularSpectrumAnalysis(F)
fig, ax = plt.subplots(1, 1, figsize=(20, 10))
ax.plot(SSA.fit(n_components=20, L='default'))
ax.plot(F)
```
