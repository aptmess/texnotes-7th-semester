```python
import itertools
import networkx as nx
import numpy.random as rnd
import matplotlib.pyplot as plt
import numpy as np
```

## 1. Complete Graph
## <font color='darkblue'> **Solution** </font> 


```python
def complete_graph(N: int) -> nx.Graph:
    '''
    N: int
        количество вершин
    '''
    graph = nx.Graph()
    N_range = range(N)
    all_pairs = itertools.permutations(N_range, 2)
    graph.add_nodes_from(N_range)
    graph.add_edges_from(all_pairs)
    return graph
```

```python
n = 5
graph = complete_graph(n)
fig, ax = plt.subplots(figsize=(8, 8))
nx.draw_circular(graph, 
         node_color='lightblue',
         node_size=750,
         with_labels=True, ax=ax)
ax.set_title(f'Complete Graph, n = {n}')
plt.show()
```

## 2. Тестный мир: Watts and Strogatz
## <font color='darkblue'> **Solution** </font> 

Сначала создается колицо из для $n$ вершин, затем каждая вершина соединяется с $k$ ближайшими соседями ($k-1$ ближайших соседей, если $k$ - нечетно)

```python
def watts_strogatz_graph(n: int, k: int) -> nx.Graph:
    """Возвращает Watts–Strogatz граф Тесного мира.

    Параметры
    ----------
    n : int
        количество вершин
    k : int
        каждая вершина соединена с k ближайшими соседями в топологическом смысле слова
    """

    if k > n:
        raise nx.NetworkXError("k > n, choose smaller k or larger n")
    if k == n:
        return complete_graph(n)

    graph = nx.Graph()
    nodes = list(range(n)) 
    for j in range(1, k // 2 + 1):
        targets = nodes[j:] + nodes[0:j]
        graph.add_edges_from(zip(nodes, targets))
    return graph
```

```python
n = 16
step = 2
for k in np.arange(2, n+step, step):
    fig, ax = plt.subplots(1, 1, figsize=(8, 8))
    graph = watts_strogatz_graph(n=n, k=k)
    nx.draw_circular(graph, 
                     node_color='lightblue',
                     node_size=250,
                     with_labels=True, 
                     ax=ax)
    ax.set_title(f'Количество соседей K = {k}')
    plt.show()
```

## 3. Случайные пуассоновские графы (Эрдеш, Рении)
## <font color='darkblue'> **Solution** </font> 

`Граф Эрдьёша-Реньи` - граф вида $G_{n, p}$, где $n$ - количество вершин графа, $p$ - вероятность того, что две вершины соединены. На само деле есть еще и граф $G_{n, m}$, где $m$ - количество рёбер графа.

```python
import random
def random_graph(n:int, p:float) -> nx.Graph:
    '''
    n: int
        количество вершин графа
    p: float
        вероятность того, что две вершины соединены
    '''
    graph = nx.Graph()
    N_range = range(n)
    graph.add_nodes_from(N_range)
    for pair in itertools.permutations(N_range, 2):
        if random.random() < p:
            graph.add_edge(*pair)

    return graph
```

```python
# colors = ['r','b','y','cyan'] * 5
for p in np.arange(0, 1.1, 0.1):
    graph = random_graph(20, p)
    fig, ax = plt.subplots(figsize=(8, 8))
    nx.draw_circular(graph,
                     node_color='lightblue',
                     edge_color = 'black',
                     node_size=1000,
                     with_labels=True, 
                     ax=ax)
    ax.set_title(f'p = {p:.2}')
    plt.show()
```

## 4. Модель Френча-Де Гроота

## <font color='darkblue'> **Solution** </font> 

$$\large x_i(k+1) = x_i(k) + f(x_1(k), x_2(k), \ldots, x_n(k))$$

Линейный случай:

$$\large x_i(k+1) = \sum\limits_{j=1}^n w_{ij}x_j(k), \forall i k = 0, 1, \ldots $$

Матрица марковской цепи

```python
peoples = 5
P = np.random.random((peoples, peoples))
P_norm = P / P.sum(axis=1).reshape(-1,1)
P_norm
```

Начальный вектор предпочтений

```python
start = np.random.randint(0, 100, peoples)
start
```

Шаги

```python
p_s = [start]
start_0 = start
for i in range(peoples-1):
    start_0 = P_norm @ start_0 
    p_s.append(start_0)
```

```python
result = np.array(p_s).T
```

Визуализация

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 8))
for person, step in enumerate(result):
    ax.plot(step, range(1, peoples+1), label=f'person {person+1}')
ax.legend()
ax.set_title('Сходимость к общему мнению')
plt.show()
```
