```python
import random
import numpy as np
import matplotlib.pyplot as plt

prob = [0.05, 0.95]  
start = 0
positions = [start]

rr = np.random.random(900)
downp = rr < prob[0]
upp = rr > prob[1]
for idownp, iupp in zip(downp, upp):
    down = idownp and positions[-1] > -1
    up = iupp and positions[-1] < 2
    positions.append(positions[-1] - down + up)

fig, ax = plt.subplots(1, 1, figsize=(16, 10))
ax.plot(positions, color='darkblue', linestyle='--')
ax.set_title('Одномерное случайное блуждание')
plt.show()
```

```python
import numpy
import random

n = 100000
x = numpy.zeros(n)
y = numpy.zeros(n)

for i in range(1, n):
    val = random.randint(1, 4)
    if val == 1:
        x[i] = x[i - 1] + 1
        y[i] = y[i - 1]
    elif val == 2:
        x[i] = x[i - 1] - 1
        y[i] = y[i - 1]
    elif val == 3:
        x[i] = x[i - 1]
        y[i] = y[i - 1] + 1
    else:
        x[i] = x[i - 1]
        y[i] = y[i - 1] - 1

fig, ax = plt.subplots(1, 1, figsize=(16, 10))
ax.plot(x, y, color='darkblue')
ax.grid()
ax.set_title('Двумерное случайное блуждание')
plt.show()
```
