# Наивный байесовский предсказатель следующего слова

```python
import numpy as np
import pandas as pd
```

## Словарики

```python
with open('count_1w.txt', 'r') as file:
    lines = file.readlines()
```

```python
words = []
nu = []
for line in lines:
    key, value = line.split('\t')
    words.append(key)
    nu.append(int(value))
    
vocabulary_1 = pd.DataFrame()
vocabulary_1['nu'] = nu
vocabulary_1.index = words
vocabulary_1['log P(w_k)'] = np.log(vocabulary_1['nu'] / vocabulary_1['nu'].sum())
vocabulary_1.head()
```

```python
with open('count_2w.txt', 'r') as file:
    lines = file.readlines()
```

```python
vocabulary_2 = {'k - 1': [], 'k' : [], 'nu' : []}
for line in lines:
    w1, w2, value = line.split()
    vocabulary_2['k - 1'].append(w1)
    vocabulary_2['k'].append(w2)
    vocabulary_2['nu'].append(int(value))
    
vocabulary_2 = pd.DataFrame(vocabulary_2)
vocabulary_2['log P(w_k-1|w_k)'] = np.log(vocabulary_2['nu'] / vocabulary_2['nu'].sum())
vocabulary_2.head()
```

## Текст для экспериментов

```python
with open('top100anecdotov.txt', 'r', encoding='utf-8') as file:
    lines = file.readlines()
```

```python
print(''.join(lines[2:5]))
```

```python
for i in range(len(lines)):
    if lines[i][:5] == '* * *':
        lines[i] = ''
        
text = ''.join(lines[7:])
print(text)
```

```python
def get_vocabularies(text):
    lst = text.lower().split()
    words = {}
    for word in lst:
        if word not in words:
            words[word] = lst.count(word)

    voc_1 = pd.DataFrame()
    voc_1['nu'] = list(words.values())
    voc_1.index = list(words.keys())
    voc_1['log P(w_k)'] = np.log(voc_1['nu'] / voc_1['nu'].sum())
    
    bigrams = [[lst[i], lst[i + 1]] for i in range(len(lst) - 1)]
    values = []
    for bigram in bigrams:
        values.append(bigram + [bigrams.count(bigram)])

    voc_2 = pd.DataFrame(values, columns=['k - 1', 'k', 'nu']).drop_duplicates()
    voc_2['log P(w_k-1|w_k)'] = np.log(voc_2['nu'] / voc_2['nu'].sum())
    
    return voc_1, voc_2
```

```python
voc_1, voc_2 = get_vocabularies(text)
voc_1.head()
```

```python
voc_2.head()
```

## Задание 1

```python
def predict_next_word(vocabulary_1, vocabulary_2, word):
    candidates = vocabulary_2[vocabulary_2['k - 1'] == word]
    candidates = vocabulary_1.loc[candidates['k'], 'log P(w_k)'] + candidates['log P(w_k-1|w_k)'].values
    if len(candidates) != 0:
        return candidates.index[np.argmax(candidates)]
    else:
        return 'the'
```

### Развлекаемся

Просто слова

```python
predict_next_word(vocabulary_1, vocabulary_2, 'back')
```

```python
predict_next_word(vocabulary_1, vocabulary_2, 'will')
```

```python
predict_next_word(vocabulary_1, vocabulary_2, 'love')
```

```python
predict_next_word(vocabulary_1, vocabulary_2, 'crying')
```

Попробуем написать предложение

```python
word = 'text'

for i in range(11):
    print(word, end=' ')
    word = predict_next_word(voc_1, voc_2, word)
    
print(word + '!')
```

##### Поработаем с текстом и импортированными словарями

```python
print('Слов в тексте:', len(text.split()))
```

```python
counter = 0
for i in range(len(text.split()) - 1):
    if text[i + 1] == predict_next_word(vocabulary_1, vocabulary_2, text[i]):
        counter += 1
```

```python
print('Угадано слов: ', counter)
```
