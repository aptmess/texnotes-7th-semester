3. На каждом шаге алгоритма выбирайте $\alpha_t$ путем решения задачи одномерной оптимизации (функция `minimize` в `scipy`).

4. Как менялось качество предсказаний на каждом шаге алгоритма? Выведите график зависимости значения среднеквадратичной ошибки от номера итерации

## Алгоритм градиентного бустинга



## Реализация

```python
def fit_minimal(alpha, y, a, _dtc, X):
    return np.sum((y - (a + alpha * _dtc.predict(X)))**2) / 2

def gradient_boosting(X, y, max_iter=10):
    b = DecisionTreeRegressor(random_state=42, max_depth=2)
    a = lambda X: np.mean(y)
    d = {}
    for i in range(max_iter):
        a_0 = a(X)
        r_i = y - a_0
        b.fit(X, r_i)
        res = minimize(fit_minimal, 
                       x0=1, 
                       args = (y, a_0, b, X))
        d[f'step{i}'] = {'alpha': res.x[0], 
                         'L': res.fun}
        a = lambda X:  a_0 + res.x * b.predict(X)
    return a, d
```

```python
algorithm, dictionary = gradient_boosting(X, y, max_iter=4)
d = pd.DataFrame(dictionary).T
fig, ax = plt.subplots(1, 2, figsize=(16, 4))
ax[0].plot(algorithm(X), label='Gradient Boosting', color='darkred')
ax[0].plot(y, label='Y True', color='darkblue')
ax[0].set_title('Исходные данные')
ax[0].legend()
d.L.plot(ax=ax[1], label='L', color='darkblue')
ax[1].legend()
ax[1].set_title('Функция потерь L')
plt.show()
d
```
