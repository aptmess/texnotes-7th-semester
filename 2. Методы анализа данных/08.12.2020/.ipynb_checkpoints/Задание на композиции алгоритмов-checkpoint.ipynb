# Построение композиций алгоритмов

**Задание 1.** Загрузите набор данных claims.csv. Для обучения моделей возьмите только вещественные признаки, пропуски замените нулями. Разбейте выборку на обучение и контроль (70% и 30%).

```python
import pandas as pd
claims = pd.read_csv('claims.csv')
```

```python
claims.head()
```

```python
preprocess = claims.select_dtypes(include=['float64', 'int64']).fillna(0)
preprocess
```

```python
X = preprocess.drop(['ID', 'target'], axis=1).values
X.shape
```

```python
y = preprocess['target'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

**Задание 2.** Для решения данной задачи обучите следующие алгоритмы:
- DecisionTreeClassifier;
- BaggingClassifier (n_estimators=100);
- RandomForestClassifier (n_estimators=100).

Оцените качество перечисленных алгоритмов на кросс-валидации по 3 блокам. Метрикой качества выберите долю верных ответов алгоритма (accuracy). Сравните значения метрики на каждом из блоков. Для сравнения используйте несколько различных `random_state` в каждом из алгоритмов. Какие выводы можно сделать?

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import accuracy_score
import numpy as np
```

```python
results = {}
cl = KFold(n_splits=3)
```

```python
for splt, (train_index, test_index) in enumerate(cl.split(X), 1):
    results[f'fold {splt}'] = {'DecisionTreeClassifier': [], 
                              'BaggingClassifier(100)': [],
                              'RandomForestClassifier(100)': []}
    print('Fitting')
    for _ in range(3):
        dtc = DecisionTreeClassifier(random_state=np.random.randint(0, 1000))
        bcl = BaggingClassifier(n_estimators=100, random_state=np.random.randint(0, 1000))
        rfc = RandomForestClassifier(n_estimators=100, random_state=np.random.randint(0, 1000))
        dtc.fit(X[train_index], y[train_index])
        bcl.fit(X[train_index], y[train_index])
        rfc.fit(X[train_index], y[train_index])
        results[f'fold {splt}']['DecisionTreeClassifier'].append(accuracy_score(dtc.predict(X[test_index]), y[test_index]))
        results[f'fold {splt}']['BaggingClassifier(100)'].append(accuracy_score(bcl.predict(X[test_index]), y[test_index]))
        results[f'fold {splt}']['RandomForestClassifier(100)'].append(accuracy_score(rfc.predict(X[test_index]), y[test_index]))
    print(f'Done Fitting fold {splt}')
```

```python
results
```

**Вывод**: `RandomForestClassifier(n_estimators = 100)` работает лучше всего

**Задание 3.** Выберем число деревьев для случайного леса. Обучите `RandomForestClassifier` с числом деревьев от 20 до 1000 с шагом 20 и оцените качество полученных алгоритмов на тестовой выборке.

Если каждый раз обучать `RandomForestClassifier` с нуля, придется обучить в общей сумме $20 + 40 + \ldots + 1000$ деревьев.
Однако деревья в случайных лесах строятся независимо и параллельно, поэтому можно обучить всего 1000 деревьев.

Для этого при создании объекта класса `RandomForestClassifier` нужно указать `warm_start=True`. Затем обучить алгоритм с помощью метода `fit`, использовать метод `predict` для классификации. После этого с помощью метода `set_params` изменить параметр `n_estimators`. Если к полученному объекту применить метод `fit`, внутри него будет обучаться только недостающее число деревьев.

Переобучается ли случайный лес с увеличением числа деревьев?

```python
result_for_trees = {}
for n in np.arange(20, 1020, 20):
    rfc_choose = RandomForestClassifier(n_estimators=100, 
                                        random_state=np.random.randint(0, 1000))
    rfc_choose.fit(X_train, y_train)
    acc = accuracy_score(rfc_choose.predict(X_test), y_test)
    result_for_trees[n] = acc
    print(f'Trees : {n}, accuracy {acc}')
```

А это, если использовать `warm_start`. Гораздо быстрее обучалось

```python
warm_start_trees = RandomForestClassifier(n_estimators=20, warm_start=True, 
                                        random_state=np.random.randint(0, 1000))
result_for_trees_ws = {}
for n in np.arange(20, 1020, 20):
    warm_start_trees.set_params(n_estimators=n)
    warm_start_trees.fit(X_train, y_train)
    acc = accuracy_score(warm_start_trees.predict(X_test), y_test)
    result_for_trees_ws[n] = acc
    print(f'Trees : {n}, accuracy {acc}')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
fig, ax = plt.subplots(1, 1, figsize=(16, 10))
ax.plot(list(result_for_trees_ws.values()), 
        color='darkblue', 
        label='Accuracy')
ax.legend()
ax.set_xlabel('n_estimators')
ax.set_title('RandomForestClassifier(n_estimators)')
plt.show()
```

**Вывод**: идёт переобучение с увеличением числа деревьев.

**Задание 5.** В первом и последующих заданиях не использовались некоторые категориальные признаки. Добавьте категориальные признаки с количеством уникальных значений < 50 тремя способами:
- как OHE признаки;
- как порядковые признаки (LabelEncoder);
- как счетчики (FrequencyEncoder).

Подберите в каждом из случаев оптимальные параметры xgboost. Изменится ли оптимальная глубина деревьев от способа кодирования категориальных признаков? Согласуется ли это с экспериментом? Как меняется время, необходимое для обучения модели в зависимости от способа кодирования?

```python
categorical_features = claims.drop(preprocess.columns, axis=1)
categorical_features
```

Найдём категориальные признаки с количество уникальных значений <50 тремя способами.

```python
ct_unique = {}
for i in categorical_features.columns:
    ct_unique[i] = pd.unique(categorical_features[i]).shape[0]
```

```python
ct_less_50_features = {k:v for (k, v) in ct_unique.items() if v < 50}
ct_less_50_features
```

**5.1** `OneHotEncoder`

```python
ohe = pd.get_dummies(categorical_features[ct_less_50_features.keys()].fillna(0))
print(ohe.shape)
ohe.head()
```

```python
X_ohe = pd.concat((preprocess, ohe), axis=1).drop(['ID', 'target'], axis=1).values
```

```python
X_train_ohe, X_test_ohe, y_train_ohe, y_test_ohe = train_test_split(X_ohe, y, test_size=0.3, random_state=42)
```

```python
import time
result_dict = {}
xgb_ohe = xgboost.XGBClassifier()
cl_k = KFold(n_splits=3, shuffle=True, random_state=0)
params = {'learning_rate': np.arange(0.05, 0.35, 0.05), 
          'n_estimators': np.arange(50, 200, 50), 
          'max_depth': np.arange(1, 10, 3)}
start = time.time()
best_xgb = GridSearchCV(estimator=xgb, 
                        param_grid=params,
                        cv=cl_k)
best_xgb.fit(X_train_ohe, y_train_ohe)
```

```python
result_dict['OHE'] = {'acc': accuracy_score(y_test_ohe, best_xgb.predict(X_test_ohe)), 'time': time.time() - start}
```

```python
result_dict
```

```python
best_xgb.best_params_
```

5.2 **Label Encoder**

```python
from sklearn.preprocessing import LabelEncoder
```

```python
step_1_label = categorical_features[ct_less_50_features.keys()].fillna(0)
step_1_label
```

```python
xle = preprocess.copy()
```

```python
for i in step_1_label.columns:
    xle = pd.concat(
        (xle, 
         pd.Series
         (
             LabelEncoder().fit_transform(step_1_label[i].astype(str)), 
             name=i
         )
        ), 
        axis=1)
```

```python
xle
```

```python
X_label_encoder = xle.drop(['ID', 'target'], axis=1).values
```

```python
X_train_label, X_test_label, y_train_label, y_test_label = train_test_split(X_label_encoder, y, test_size=0.3, random_state=42)
```

```python
xgb_label = xgboost.XGBClassifier(use_label_encoder=False)
cl_k = KFold(n_splits=3, shuffle=True, random_state=0)
params = {'learning_rate': np.arange(0.05, 0.35, 0.05), 
          'n_estimators': np.arange(50, 200, 50), 
          'max_depth': np.arange(1, 10, 3)}
start = time.time()
best_xgb = GridSearchCV(estimator=xgb_label, 
                        param_grid=params,
                        cv=cl_k)
best_xgb.fit(X_train_label, y_train_label)
```

```python
best_xgb.best_params_
```

```python
result_dict['Label'] = {'acc': accuracy_score(y_test_label, best_xgb.predict(X_test_label)), 
                        'time': time.time() - start}
```

```python
result_dict
```

**Вывод**: в зависимости от кодирования не меняется максимальная глубина, точность одинаковая, только `label encoder` работает быстрее.
