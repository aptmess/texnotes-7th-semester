## 1.1 Исходные данные




Разберёмся с исходными данными. 

- `name` - уникальный идентификатор города;
- `Latitude` - *широта*, идёт с севера (`N`) на юг (`S`), измеряется в градусах, лежит в интервале $\operatorname{Lat} \in [-90, 90]$
    - $\operatorname{Lat} = 0 $ - экватор
    - $\operatorname{Lat} <0 $ - южное полушарие
    - $\operatorname{Lat} >0 $ - сереверное полушарие
- `Longitude` - *долгота*, идёт с запада (`W`) на восток (`E`), измеряется в градусах, лежит в интервале $\operatorname{Long} \in [-180, 180]$
    - $\operatorname{Long} = 0 $ - основной меридиан (через Гринвич проходящий)
    - $\operatorname{Long} < 0 $ - западное полушарие
    - $\operatorname{Long} > 0 $ - восточное полушарие

- `Weight(g)` - вес подарка в граммах (`g`)

Координаты каждого города на *Земле (Earth)* задаются с помощью пары `Latitude, Longitude`.

## 1.3 Расстояние между двумя городами на сфере

Так как у нас предполагается, что города находятся на планете *Земля*, которая в свою очередь является сферой, то обычное евклидово расстояние не способно вычислить реальное расстояние между городами. Для подсчёта расстояния используется формула [`great_circle`](https://en.wikipedia.org/wiki/Great-circle_distance).

Реализация (сравнение так же можно посмотреть) взята из данного [файла](https://gist.github.com/habibutsu/8bbcc202a915e965c6a6d4f561d0e482).

```python
from math import radians, sin, cos, asin, sqrt, pi, atan, atan2, fabs
def great_circle(point1, point2, EARTH_MEAN_RADIUS=6371008.8):
    '''
        Calculating great-circle distance
        (see https://en.wikipedia.org/wiki/Great-circle_distance)
    '''
    lat1, lon1 = (radians(coord) for coord in point1)
    lat2, lon2 = (radians(coord) for coord in point2)

    dlon = fabs(lon1 - lon2)
    dlat = fabs(lat1 - lat2)

    numerator = sqrt(
        (cos(lat2) * sin(dlon))**2 +
        ((cos(lat1) * sin(lat2)) - (sin(lat1) * cos(lat2) * cos(dlon)))**2)

    denominator = (
        (sin(lat1) * sin(lat2)) +
        (cos(lat1) * cos(lat2) * cos(dlon)))

    c = atan2(numerator, denominator)
    return EARTH_MEAN_RADIUS * c / 1000
```

Протестируем, как работает данная функция. От двух координат точек на сфере `(Latitude_1, Longitude_1)` и `(Latitude_2, Longitude_2)` возвращается длина в километрах между данными городами.

```python
p_minsk = (53.902257, 27.561831)
p_moscow = (55.75396, 37.620393)
```

## 1.4 Матрица расстояний

Высчитать матрицу расстояний, $10^6$ раз посчитав расстояние `great_circle` - неудобно и времязатратно, да и хранить такую матрицу тяжело. Создадим класс, соединяющий две вершины городов (экземпляры класса `Wishes`) и сохраняющий информацию о расстоянии между ними.

```python
class Edges:
    '''
    Edges: данный класс соединяет две вершины между собой и высчитывает расстояние между ними
    '''
    def __init__(self, v1, v2):
        '''
        self.e : tuple
            tuple из двух экземпляров класса
        self.dist : float
            расстояние в километрах между городами
            используется функция great_circle(point_1=(lat_1, long_1), 
                                              point_2=(lat_2, long_2))
        '''
        self.e = (v1, v2) if v1.wish_id < v2.wish_id else (v2, v1)
        self.dist = great_circle((v1.wish_lat, v1.wish_long), 
                                 (v2.wish_lat, v2.wish_long))
        
    def __str__(self):
        return str((self.e[0].wish_id, 
                    self.e[1].wish_id))
    
    def __contains__(self, wish):
        return (self.e[0].wish_id == wish.wish_id) or (self.e[1].wish_id == wish.wish_id)
```

### 1.4.1 `K`-ближайших соседей

Для дальнейшей работы алгоритма хотелось бы найти для каждого из городов его ближайших соседей, чтобы выбирать минимальный по расстоянию маршрут. В общем случае это сделать, наверное, тяжеловато.

Но так как у нас была построена кластеризация с помощью `K-means` на евклидовом расстоянии, то реализовав удобным образом функцию для подсчёта `euclidean_distance` мы можем применить эту функцию и найти ближайших соседей.

```python
import numpy as np


def euclidean_distance(x, y):
    '''

    Parameters
    ----------
    x : np.ndarray размера N x D
        обучающая выборка.
    y : np.ndarray размера M x D
        тестовая выборка.

    Raises
    ------
    NotImplementedError
        DESCRIPTION.

    Returns
    -------
    Функция возвращает np.ndarray размера N x M,
    каждый элемент которого - евклидово расстояния между
    соответствующей парой векторов из массивов x и y.
    Уменьшим число операций с помощью трюка, раскрывая скобки.
    '''
    x2 = np.sum(x**2, axis=1).reshape(-1, 1)
    y2 = np.sum(y**2, axis=1)
    result = np.sqrt(x2 + y2 - 2 * x @ y.T)
    return result
```

```python
def get_best_ranks(ranks, top, axis=1, return_ranks=False):
    indices = np.take(np.argpartition(ranks, range(-top, 0), axis=axis),
                      range(-1, -top-1, -1), axis=axis)
    result = (indices, )
    if return_ranks:
        ranks = np.take_along_axis(ranks, indices, axis=axis)
        result = (ranks, ) + result
    return result if len(result) > 1 else result[0]
```

## 1.6 И ещё немного классов

Давайте теперь создам класс, описывающий информацию о загрузке саней в определённый момент времени, информацию о городах, которые необходимо посетить и общий вес подарков.

```python
class Sleigh:
    
    def __init__(self, max_weight=10000 * 1000):
        self.visit = []
        self.load = 0
        self.max_weight = max_weight
        
    def add(self, vertex):
        if vertex in self.visit:
            return False
        
        if self.load + vertex.wish_w <= self.max_weight:
            self.visit.append(vertex)
            self.load += vertex.wish_w
            #print('#children:', len(self.visit), ':', self.load / 1000, 'kg (', 100 * self.load / self.max_weight, '%)')
            return True
        return False

    def clear(self):
        self.visit.clear()
```

А теперь будем добавлять информацию о каждой поездке в `Tour`.

```python
class Tour:
    
    def __init__(self):
        self.edges = []
        
    def add(self, edg):
        if (edg not in self.edges):
            self.edges.append(edg)
            return True
        return False

    def get_cost(self):
        t = 0
        for e in self.edges:
            t += e.dist     
        return t

    def is_valid(self):
        if len(self.edges) == 0:
            return False
        a = []
        for edg in self.edges:
            c1, c2 = edg.e
            a.append(c1.wish_id)
            a.append(c2.wish_id)
        b = np.unique(a, return_counts = True)
        return len(b[1]) == len(np.where(b[1] == 2)[0])
```

А теперь напишем функцию, возвращающую маршрут саней.

```python
def travel(s, start):
    a = s.visit
    t = Tour()
    e = Edges(start, a[0])
    t.add(e)
    for i in range(len(a) - 1):
        e = Edges(a[i], a[i + 1])
        t.add(e)
    e = Edges(a[-1], start)
    t.add(e)
    return t 
```

## 2. Find Solution

```python
def solve(children, start_point, areas, n, GEO_AREAS, nearest_neighbours_K=5):
    Q_MAX = 70 # number of tries to build randomly a tour
    #areas = clusters(children)
    #n = read_neigbors(children) # lookup container for the best, close neighbors
    
    tally_best_visited = []  # solution IDs
    best_visited = []        # best tour for a starting child in an area
    total = 0                # distance of tour for a starting child in an area
    tally_total = 0      # solution distance
    tally_count = 0      # sum up processed IDs, for verification
    ii = 0              # just indexing geo.areas
    count = 0 # count children in tours in a geo.area
    tally_result = {k: [] for k in GEO_AREAS}

    for a in areas: # iterate geographical areas
        tally_min = 60e6
        print()
        print('Area:', GEO_AREAS[ii])
        
        k = 0
        for start in a: # iteratively pick up first child in the geo.area
            k +=1
            aux = set(a) # make an auxiliary copy of children as a set
            total = 0
            best_visited = []
            count = 0
            init = True
            first = True
            while len(aux):
                    
                s_sleigh = Sleigh()
                if init:
                    init=False
                    item = start
                    aux.remove(item)
                else:
                    item = aux.pop()
                s_sleigh.add(item)
                first = False
                
                while not first and len(aux):    
                    closest = n[item]
                    i = 0
                    ret = False
                    was_full = False
                    while i < nearest_neighbours_K:
                        item = closest[i]
                        i += 1
                        if item in aux:
                            ret = s_sleigh.add(item)
                            was_full = not ret
                            if ret:
                                aux.discard(item)
                                break
                    # 1. available, didn't fit; 2. not available
                    if ret == False and (not was_full):
                        temp = []
                        while len(aux):
                            item = aux.pop()
                            ret = s_sleigh.add(item)
                            if ret:
                                for x in temp:
                                    aux.add(x)
                                break
                            temp.append(item)
                        if not ret:
                            was_full = True
                            for x in temp:
                                aux.add(x)
                            
                    if ret == False and was_full:
                        first = True
                        count += len(s_sleigh.visit)
                        if len(s_sleigh.visit) > 2:
                            t_min = 66e6
                            q = 0
                            while q < Q_MAX:
                                q += 1
                                t_tour = travel(s_sleigh, start_point)
                                if t_tour.is_valid():
                                    cost = t_tour.get_cost()
                                    if cost < t_min:
                                        t_min = cost
                                        best = deepcopy(s_sleigh.visit)
                                np.random.shuffle(s_sleigh.visit)
                            total += t_min
                            best_visited.append(best)
                        else:        
                            t_tour = travel(s_sleigh, start_point)       
                            if t_tour.is_valid():
                                cost = t_tour.get_cost()
                                best = deepcopy(s_sleigh.visit)
                                best_visited.append(best)
                                total += cost
                        break
                    
            if not len(aux):
                count += len(s_sleigh.visit)
                if len(s_sleigh.visit) > 2:
                    t_min = 66e6
                    q = 0
                    while q < Q_MAX:
                        q += 1
                        t_tour = travel(s_sleigh, start_point)
                        if t_tour.is_valid():
                            cost = t_tour.get_cost()
                            if cost < t_min:
                                t_min = cost
                                best = deepcopy(s_sleigh.visit)
                        np.random.shuffle(s_sleigh.visit)
                    total += t_min
                    best_visited.append(best) 
                else:        
                    t_tour = travel(s_sleigh, start_point)
                    if t_tour.is_valid():
                        cost = t_tour.get_cost()
                        best = deepcopy(s_sleigh.visit)
                        best_visited.append(best)
                        total += cost
                
                if total < tally_min:
                    tally_min = total
                    tally_best = best_visited
                    print()
                    print('Area', ii, 'of', len(areas)-1, 'Best dist:', tally_min, 
                          tally_min / 1e6, 'Mill.km')
                    print('Child', k, 'of', len(a), ' Total dist:', total, 
                          total / 1e6, 'Mill.km')
        
        tally_total += tally_min
        tally_count += count
        tally_best_visited += tally_best
        tally_result[GEO_AREAS[ii]] += tally_best
        tally_result[GEO_AREAS[ii]] = [[j.wish_id for j in i] for i in tally_result[GEO_AREAS[ii]]]
        ii += 1
        print(f'Current Total Distance: {tally_total / 1e6} Mill.km')
        wish_result_ways = [[j.wish_id for j in i] for i in tally_best_visited]
        yeah = np.any([(df.iloc[np.array(i)-2]['Weight (g)'].sum() / 1000) < 10000 for i in wish_result_ways])
        try:
            print(f'Every Toure is < 10000 kg: {yeah}')
            print(ii-1)
        except:
            print('Mistake')

    print('--- Total:', tally_total, tally_total / 1e6, 'Mill.km')
    print('ID count OK?:', tally_count, tally_count == len(children))
    print()
    return tally_total, tally_best_visited, tally_result
```
